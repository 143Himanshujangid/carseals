```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Load the Dataset
file_path = "C:\\Users\\HIMANSHU\\OneDrive\\Desktop\\piyush\\UserCarData.csv"  
data = pd.read_csv(file_path)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
##What is the distribution of car selling prices?

plt.figure(figsize=(10, 6))
sns.histplot(data['selling_price'], bins=30, kde=True, color='blue')
plt.title('Distribution of Selling Prices')
plt.xlabel('Selling Price (â‚¹)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Question 2: How does mileage vary with car price?
plt.figure(figsize=(10, 6))
sns.scatterplot(x='mileage', y='selling_price', hue='fuel', data=data, palette='viridis', alpha=0.7)
plt.title('Mileage vs. Selling Price by Fuel Type')
plt.xlabel('Mileage (kmpl)')
plt.ylabel('Selling Price (â‚¹)')
plt.legend(title='Fuel Type')
plt.show()

```

```python
# Top 10 Most Common Car Brands
top_brands = data['name'].value_counts().nlargest(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_brands.index, y=top_brands.values, palette='coolwarm', hue=top_brands.index, dodge=False, legend=False)
plt.title('Top 10 Most Common Car Brands')
plt.xlabel('Car Brand')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Fuel Type Distribution
plt.figure(figsize=(8, 5))
sns.countplot(x='fuel', data=data, hue='fuel', dodge=False, palette='viridis')
plt.title('Fuel Type Distribution')
plt.xlabel('Fuel Type')
plt.ylabel('Count')
plt.show()
```

```python
# Boxplot: Mileage by Fuel Type
plt.figure(figsize=(10, 6))
sns.boxplot(x='fuel', y='mileage', data=data, hue='fuel', dodge=False, palette='pastel')
plt.title('Fuel Type vs. Mileage')
plt.xlabel('Fuel Type')
plt.ylabel('Mileage (kmpl)')
plt.show()
```

```python
# Question 6: What is the impact of car age on selling price?
data['car_age'] = 2024 - data['year']
plt.figure(figsize=(10, 6))
sns.scatterplot(x='car_age', y='selling_price', hue='fuel', data=data, palette='coolwarm', alpha=0.7)
plt.title('Car Age vs. Selling Price')
plt.xlabel('Car Age (Years)')
plt.ylabel('Selling Price (â‚¹)')
plt.show()


```

```python
# Ownership Type Distribution
owner_counts = data['owner'].value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=owner_counts.index, y=owner_counts.values, palette='cividis', hue=owner_counts.index, dodge=False, legend=False)
plt.title('Ownership Type Distribution')
plt.xlabel('Ownership Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Question 8: What is the distribution of car engine sizes?
plt.figure(figsize=(10, 6))
sns.histplot(data['engine'], bins=30, kde=True, color='orange')
plt.title('Distribution of Engine Sizes')
plt.xlabel('Engine Size (cc)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Question 9:How does engine size relate to selling price?
plt.figure(figsize=(10, 6))
sns.scatterplot(x='engine', y='selling_price', hue='fuel', data=data, palette='viridis', alpha=0.7)
plt.title('Engine Size vs. Selling Price')
plt.xlabel('Engine Size (cc)')
plt.ylabel('Selling Price (â‚¹)')
plt.show()
```

```python
# Boxplot: Transmission Type vs. Selling Price
palette = {'Manual': 'lightblue', 'Automatic': 'lightgreen'}
plt.figure(figsize=(8, 5))
sns.boxplot(x='transmission', y='selling_price', data=data, hue='transmission', dodge=False, palette=palette)
plt.title('Transmission Type vs. Selling Price')
plt.xlabel('Transmission Type')
plt.ylabel('Selling Price (â‚¹)')
plt.show()
```

```python
#12 Scatterplot: Kilometers Driven vs. Selling Price
plt.figure(figsize=(10, 6))
sns.scatterplot(x='km_driven', y='selling_price', hue='fuel', data=data, palette='coolwarm', alpha=0.7)
plt.title('Kilometers Driven vs. Selling Price')
plt.xlabel('Kilometers Driven')
plt.ylabel('Selling Price (â‚¹)')
plt.show()
```

```python
# 10. What is the relationship between max power and selling price?
plt.figure(figsize=(10, 6))
sns.scatterplot(x='max_power', y='selling_price', hue='fuel', data=data, palette='viridis', alpha=0.7)
plt.title('Max Power vs. Selling Price')
plt.xlabel('Max Power (bhp)')
plt.ylabel('Selling Price (â‚¹)')
plt.show()
```
